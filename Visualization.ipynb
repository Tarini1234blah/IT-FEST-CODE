# VISUALIZATION

Let us start with Gender


```python
Supermarket.groupby(['Gender']). agg({'Total':'sum'})
```

```python
plt.figure(figsize=(14,7))
plt.style.use('fivethirtyeight')
ax= sns.countplot('Gender', data=Supermarket , palette = 'copper')
ax.set_xlabel(xlabel= "Gender",fontsize=18)
ax.set_ylabel(ylabel = "Gender count", fontsize = 18)
ax.set_title(label = "Gender count in supermarket", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (14,6))
ax = sns.countplot(x="Gender", hue = "Product line", data = Supermarket, palette= "tab20")
ax.set_title(label = "Products sales based on Gender", fontsize= 25)
ax.set_xlabel(xlabel = "Gender", fontsize = 16)
ax.set_ylabel(ylabel = "Product line", fontsize = 16)
```

#Analysis based on Branch

```python
plt.figure(figsize = (14,6))
ax = sns.countplot(x="Branch",  hue = "Product line", data = Supermarket, palette= "rocket_r")
ax.set_title(label = "Branch wise Product sales", fontsize= 25)
ax.set_xlabel(xlabel = "Branch", fontsize = 16)
ax.set_ylabel(ylabel = "Product line", fontsize = 16)
```

```python
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =Supermarket, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Rating distribution", fontsize = 16)
```

#Customer Type

```python
Supermarket.groupby(['Customer type']). agg({'Total':'sum'})
```

```python
plt.figure(figsize= (14,6))
ax = sns.countplot(x = "Customer type", data = Supermarket, palette = "rocket_r")
ax.set_title("Type of customers", fontsize = 25)
ax.set_xlabel("Customer type", fontsize = 16)
ax.set_ylabel("Customer Count", fontsize = 16)
```

```python
plt.figure(figsize=(14,6))
ax = sns.countplot(x = "Customer type", hue = "Branch", data = Supermarket, palette= "rocket_r")
ax.set_title(label = "Customer type in different branch", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Customer Count", fontsize = 16)
```

```python
plt.figure(figsize=(14,6))
ax = sns.lineplot(x = "Customer type", y = "Total", data = Supermarket, palette= 'rocket_r')
ax.set_xlabel(xlabel = "Customer type", fontsize = 16)
ax.set_ylabel(ylabel = "Total", fontsize = 16)
```

#Product Line Analysis

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(x= "Product line", y= "Total", data = Supermarket)
ax.set_title(label = " Total sales of product", fontsize = 25)
ax.set_xlabel(xlabel = "Product Line", fontsize = 16)
ax.set_ylabel(ylabel = "Total Sales", fontsize = 16)
```

```python
plt.figure(figsize = (14,6))
ax= sns.barplot(y= "Product line", x = "Total", hue = "Gender", data = Supermarket)
ax.set_title(label = "Product sales on the basis of gender")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Product Line")
```

```python
plt.figure(figsize = (14,6))
sns.relplot(y = 'Product line', x = 'gross income', data=Supermarket )
```

```python
plt.figure(figsize = (14,6))
ax = sns.boxplot(y = "Product line", x = "Rating", data = Supermarket)
ax.set_title("Average rating of product line", fontsize = 25)
ax.set_xlabel("Rating", fontsize = 16)
ax.set_ylabel("Product line", fontsize = 16)
```

#Quantity Analysis

```python
plt.figure(figsize=(14,6))
sns.lineplot(x="Payment",  y = "Quantity", data = Supermarket)
ax.set_title(label = "Payment method based on Quantity", fontsize = 25)
ax.set_xlabel(xlabel = "Quantity", fontsize = 16)
ax.set_ylabel(ylabel= "Payment", fontsize = 16)
```
